## Visualizing Data

```python
import matplotlib.pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
GDP = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
plt.plot(years, GDP, color='g', marker='o', linestyle='solid')
plt.title('Nominal GDP')
plt.xlabel('Years')
plt.ylabel('Billions ($)')
plt.tight_layout() # it removed --> [<matplotlib.lines.Line2D at 0x2867f21a710>]
```

```python
plt.plot(years, GDP, color='m', marker='o', linestyle='--')
plt.title('Nominal GDP')
plt.xlabel('Years')
plt.ylabel('Billions ($)')
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

# Plot
plt.plot(x, y, marker='o')
plt.title("Annotated Plot")

# Annotate the highest point
plt.annotate(
    "Highest Point", 
    xy=(5, 50),         # Coordinates of the point
    xytext=(3, 45),     # Coordinates for the annotation text
    arrowprops=dict(arrowstyle="->")  # Add an arrow
)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [60, 62, 64, 64, 64, 65, 68, 69, 70, 72]
y = [102, 115, 120, 128, 135, 140, 175, 200, 210, 220]
labels = ['g', 'd', 'f', 'h', 'b', 'i', 'a', 'c', 'e']

# Create scatter plot
plt.scatter(x, y, color="navy")

# Add labels to the points
for i, label in enumerate(labels):
    plt.text(x[i], y[i], label, fontsize=9, ha='left', va='bottom')

# Add title and labels
plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("Daily minutes spent on the site")

# Display the plot
plt.show()
```
