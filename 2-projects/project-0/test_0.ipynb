## Machine Learning

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
df = pd.read_csv('area_prices.csv')
```

```python
df
```

```python
plt.scatter(df['area'], df['price'], marker='*', color='r')
plt.xlabel('Area (m^2)')
plt.ylabel('Prices ($)')
plt.title('Area vs. Prices of Homes')
plt.tight_layout()
```

```python
model = linear_model.LinearRegression()
```

```python
df[['area']]
```

```python
model.fit(df[['area']], df.price)
```
