```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

- We will be predicting the income per capita in 2020

```python
plt.scatter(df['year'], df['per capita income (US$)'], marker='*')
plt.xlabel('Years')
plt.ylabel('Income Per Capita (US$)')
plt.title("Canada's income per capita for each year")
plt.tight_layout()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
prediction = reg.predict([[2020]])
```

```python
plt.scatter(df['year'], df['per capita income (US$)'], marker='*')
plt.scatter(2020, prediction, marker='*', color='red')
plt.xlabel('Years')
plt.ylabel('Income Per Capita (US$)')
plt.title("Canada's income per capita for each year")
plt.tight_layout()
```

```python
f'Thus the Canada income per capita in 2020 is ${float(prediction)}'
```
