```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
```

```python
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.plot(x, y, 'g>')
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.title('Weather')
plt.tight_layout()
```

```python
plt.plot(x, y, 'g+')
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.title('Weather')
plt.tight_layout()
```

```python
plt.plot(x, y, linestyle='dotted', linewidth=5, color='pink')
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.title('Weather')
plt.tight_layout()
```

```python
plt.plot(x, y, 'r*--')
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.title('Weather')
plt.tight_layout()
```

```python
plt.plot(x, y, color='green', marker='+', markersize='20', linestyle='')
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.title('Weather')
plt.tight_layout()
```

```python
# alpha will control the transparency
plt.plot(x, y, color='green', alpha=0.3)
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.title('Weather')
plt.tight_layout()
```
