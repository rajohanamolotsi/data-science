```python
import matplotlib.pyplot as plt
```

```python
values = [1400, 600, 300, 410, 250]
value_labels = ['Home Rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other Utilities']
```

```python
plt.pie(values, labels=value_labels)
# plt.axis('equal')
plt.tight_layout()
```

```python
plt.pie(values, labels=value_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0.1, 0, 0, 0, 0])
plt.tight_layout()
```

```python
plt.pie(values, labels=value_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0.1, 0, 0.15, 0.25, 0.17], startangle=180)
plt.tight_layout()
```
