```python
import matplotlib.pyplot as plt
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
normal = pre_diabetic = diabetic = 0
for idx in blood_sugar:
    if idx >= 80 and idx < 100:
        normal += 1
    elif idx >= 100 and idx < 125:
        pre_diabetic += 1
    elif idx >= 125:
        diabetic += 1
```

```python
print(normal, pre_diabetic, diabetic)
```

```python
plt.hist([normal, pre_diabetic, diabetic])
plt.tight_layout()
```

```python
plt.hist(blood_sugar, bins=3, rwidth=.95)
plt.xlabel('Number of Patients')
plt.ylabel('Sugar Levels')
plt.tight_layout()
```

```python
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=.95)
plt.xlabel('Number of Patients')
plt.ylabel('Sugar Levels')
plt.tight_layout()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=.95, color=['blue', 'orange'], label=['men', 'women'])
plt.xlabel('Number of Patients')
plt.ylabel('Sugar Levels')
plt.title('Patents vs. Sugar Levels')
plt.legend()
plt.tight_layout()
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=.95, color=['blue', 'orange'], label=['men', 'women'], orientation='horizontal')
plt.xlabel('Number of Patients')
plt.ylabel('Sugar Levels')
plt.title('Patents vs. Sugar Levels')
plt.legend()
plt.tight_layout()
```
