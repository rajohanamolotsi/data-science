### Data Science

```python
users = [
    {'id': 0, 'name': 'Hero'},
    {'id': 1, 'name': 'Dunn'},
    {'id': 2, 'name': 'Sue'},
    {'id': 3, 'name': 'Chi'},
    {'id': 4, 'name': 'Thor'},
    {'id': 5, 'name': 'Clive'},
    {'id': 6, 'name': 'Hicks'},
    {'id': 7, 'name': 'Devin'},
    {'id': 8, 'name': 'Kate'},
    {'id': 9, 'name': 'Klein'}
]
```

```python
users
```

```python
friendships = [
    (0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
    (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)
]
```

```python
friendships
```

```python
for user in users:
    user['friends'] = []
```

```python
users
```

```python
for i, j in friendships:
    users[i]['friends'].append(users[j])
    users[j]['friends'].append(users[i])
```

```python
users
```

```python
users[9]
```

```python
def number_of_friends(user):
    return len(user['friends'])
```

```python
total_connections = sum(number_of_friends(user)
                        for user in users)
```

```python
total_connections
```

```python
from __future__ import division
num_users = len(users)
avg_connections = total_connections / num_users
```
